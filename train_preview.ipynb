```python
import matplotlib.pyplot as plt
import re

with open("./run_5.log") as fin:
    logs = fin.read()
    loss_hist = re.findall(r"\{'loss': ([\d\.]+),", logs)
    loss_hist = list(map(float, loss_hist))

plt.plot(loss_hist)
plt.xlabel("step")
plt.ylabel("loss")
plt.title("Training loss over time")

plt.show()
```
